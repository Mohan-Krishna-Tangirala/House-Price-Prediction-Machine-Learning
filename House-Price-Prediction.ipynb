#### Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

#### Importing the Boston House Price Dataset

```python
price_dataset = sklearn.datasets.load_boston()
```

```python
price_dataset
```

```python
# Loading the dataset to pandas DataFrame

price_df = pd.DataFrame(price_dataset.data, columns = price_dataset.feature_names)
```

```python
price_df
```

```python
# add the target column to the DataFrame

price_df['price'] = price_dataset.target
```

```python
price_df
```

```python
price_df.shape
```

```python
price_df.isnull().sum()
```

```python
price_df.describe()
```

#### Understanding the correlation between various features in the dataset

```python
corr = price_df.corr()
```

```python
corr
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(corr, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size': 14}, cmap = 'Greens')
plt.show()
```

#### Splitting the dataset into data and target

```python
X = price_df.drop(columns = 'price', axis = 1)
Y = price_df['price']
```

```python
X
```

```python
Y
```

#### Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model Training - XGBoost Regressor


```python
# loading the model

model = XGBRegressor()
```

```python
# training the model with training data

model.fit(X_train, Y_train)
```

#### Evaluation

```python
# Accuracy for prediction on training data

training_data_prediction = model.predict(X_train)


training_data_prediction
```

```python
# R squared error

score_1 = metrics.r2_score(Y_train, training_data_prediction)


# Mean Absolute Error

score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
score_1
```

```python
score_2
```

```python
# Accuracy for prediction on test data

test_data_prediction = model.predict(X_test)


test_data_prediction
```

```python
# R squared error

score_1 = metrics.r2_score(Y_test, test_data_prediction)


# Mean Absolute Error

score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
score_1
```

```python
score_2
```
